# Compartment 1: Parsing Safety Data Sheets from .pdf format
Safety Data Sheet mostly come in .pdf format from suppliers repositories/websites. Visual appearance differs, but the content back bone stays the same - therefore, we aim to extract the plain text from the .pdf to automate search and filling in forms.


```python
pip install PyPDF2
```

PdfReader is a package for parsing (extracting) text and meta data from .pdf files.

```python
from PyPDF2 import PdfReader
import os

# Check if file exists and print current working directory for debugging
print("Current working directory:", os.getcwd())
print("Files in current directory:", os.listdir())

# Update the path to your PDF file if it's in a different location
pdf_path = "50-76-0_Actinomycin D_EN(2024).pdf"
if os.path.exists(pdf_path):
	reader = PdfReader(pdf_path)
	meta = reader.metadata
	page_content = ""  # Initialize page_content variable for later use
else:
	print(f"File not found: {pdf_path}")
	print("Please check the file path and ensure the PDF is in the correct location.")
```

In the following, we extract general information from the SDS (i.e. author, creator software, length of document,...). Most valuable information in the length of the document, which later serves as iterator.

```python
print("Total Pages: ", len(reader.pages))
# All of the following could be None!
print("Author: ", meta.author)
print("Creator: ", meta.creator)
print("Producer: ", meta.producer)
print("Subject: ", meta.subject)
print("Title: ", meta.title)
```

This cell is the actual parser: for n we use. the length of the document (see above). It then reads the string (characters incl. spaces, dots and commata) pagewise from the pdf and sums them up. Finally we print the page_content to see the successful operation - this is optional and can be erased later.

```python
#Funktion ist anscheinend abhängig von der genannten Seite bei page=read_pdf.pages[]
#Um die Bindung an die Seitenzahl zu umgehen, wurde ein Loop eingefügt. Dieser Loop führt dazu,
#dass für alle reader.pages Argumente i bis n der Seiteninhalt als Text extrahiert und ausgegeben wird.
n = len(reader.pages)
for i in range(0,n):
    with open("50-76-0_Actinomycin D_EN(2024).pdf", "rb") as pdf_file:
        read_pdf = PdfReader(pdf_file)
        number_of_pages = len(read_pdf.pages)
        page = read_pdf.pages[i]
        page_content += page.extract_text()
    print(page_content)
```

Through the following function we aim to strip whitespace characters from the end of each line as the negatively impact our search via regular expression.

```python
for i in range (0,n): 
    page_content.rstrip()
```

# Compartment 2: Search for Key Structures via Regular Expression
In the next compartment we aimed to define unique key structures (type:string) and search them in the given document. As key structures might be used multiple times in the document, we matched them against themselves and only print them once. Printing the structure again serves as a control mechanism. 

```python
#search special patterns in the SDS doc
#example for patterns: CAS number, Hazard statements, Precautionary statements, etc.
#test: search via regular expression

import re
#definition of the CAS pattern plus search for it in the page content
cas_number = re.compile(r'\b\d{2,7}-\d{2}-\d{1}\b')
cas_match = re.findall(cas_number, page_content)
printed_cas = set()
#loop to print the found CAS numbers, but only if they have not been printed before (to avoid duplicates)
for cas_match in re.finditer(cas_number, page_content):
    event_cas = cas_match.group(0)
    if event_cas not in printed_cas:
        print(event_cas)
        printed_cas.add(event_cas)  

hazard_statements = re.compile(r'(H[2-4]\d{2}\s{2}.*)')
hazard_match = re.findall(hazard_statements, page_content)
printed_hazard_statements = set()
for hazard_match in re.finditer(hazard_statements, page_content):
    event_hazard_statements = hazard_match.group(0)
    if event_hazard_statements not in printed_hazard_statements:
        print(event_hazard_statements)
        printed_hazard_statements.add(event_hazard_statements)  

precautionary_statements = re.compile(r'(P[2-5]\d{2}.*)')
precautionary_match = re.findall(precautionary_statements, page_content)
printed_precautionary_statements = set()
for precautionary_match in re.finditer(precautionary_statements, page_content):
    event_precautionary_statements = precautionary_match.group(0)
    if event_precautionary_statements not in printed_precautionary_statements:
        print(event_precautionary_statements)
        printed_precautionary_statements.add(event_precautionary_statements)    
        
```
